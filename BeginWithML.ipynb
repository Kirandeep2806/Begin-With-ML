```python
data["HeartDisease"] = data["HeartDisease"].map({1: "Yes", 0:"No"})
```

```python
def convertToInt(boolValue) -> int:
  if boolValue == "Yes":
    return 1
  if boolValue == "No":
    return 0

data["HeartDisease"] = data["HeartDisease"].map(convertToInt) # Effects on assigned value
```

```python
data
```

```python
data.describe()
```

```python
def divideByAge(row):
    if row[0] >= 28 and row[0] <40:
        return "Low age"
    if row[0] >=40 and row[0]<61:
        return "Ok"
    if row[0]>60:
        return "High age"

data["Age factor"]=data[["Age"]].apply(divideByAge, axis=1)
```

```python
q1 = data["Cholesterol"].quantile(.25)
q1
```

```python
q2 = data["Cholesterol"].quantile(.5)
q2
```

```python
q3 = data["Cholesterol"].quantile(.75)
q3
```

```python
interQuantileRange = q3 - q1
interQuantileRange
```

```python
lowerLimit = q1 - (1.5*interQuantileRange)
lowerLimit
```

```python
upperLimit = q3 + (1.5*interQuantileRange)
upperLimit
```

```python
# Let's find how many outliers are present

def getOutliers(rowName):
  if rowName[0] > q3:
    return True
  if rowName[0] < q1:
    return True
  else:
    return False

temp = data[["Cholesterol"]].apply(getOutliers, axis=1)
```

```python
# Series is 1-D array so we need to choose the values that are True and we can get it by

outlierCount = len(temp[temp == True])
print(outlierCount)

# From the Series temp we are selecting each element whose value is True and counting the total length
```

```python
# Treating the outliers

def limit_imputer(value):
  if value < lowerLimit:
    return lowerLimit
  elif value > upperLimit:
    return upperLimit
  else:
    return value
```

```python
data["Cholesterol"] = data["Cholesterol"].apply(limit_imputer)
data["Cholesterol"].max()
```

```python
# Replacing NaN values with some condition
data.info()
# Still RestingBP, Cholesterol and HeartDisease are having NaN values. So, lets replace them with some condition.
```

```python
from sklearn.impute import SimpleImputer
```

```python
# Fills the NaN with the mean values of respective non-NaN cols

impute = SimpleImputer(missing_values=np.nan, strategy="mean")
X = impute.fit_transform(([[1, np.nan, np.nan], [np.nan, np.nan, np.nan], [4, 5, np.nan]]))
X
```

```python
# Lets apply the above one to the actual dataframe

# SimpleImputer expects 2D array but data["Cholesterol"] is a 1D Series object, so do
# 1D Series -> numpy array -> Reshape to 2D array

cholesterolValues = data["Cholesterol"].values.reshape(-1, 1) # 1D Series -> numpy array(using .values) -> Reshape to 2D array(using .reshape)
impute = SimpleImputer(missing_values=np.nan, strategy="most_frequent")
data["Cholesterol"] = impute.fit_transform(cholesterolValues)
```

```python
# Now lets create 10 bins
# cut() in pandas is used to segregate the data in the bins

data["Age_Group_Cholesterol"] = pd.cut(data["Cholesterol"], bins=10, labels=["Age_Group_Cholesterol_0", "Age_Group_Cholesterol_1", "Age_Group_Cholesterol_2", "Age_Group_Cholesterol_3", "Age_Group_Cholesterol_4", "Age_Group_Cholesterol_5", "Age_Group_Cholesterol_6", "Age_Group_Cholesterol_7", "Age_Group_Cholesterol_8", "Age_Group_Cholesterol_9"], include_lowest=True)

# include_lowest=True make the first interval of te bin as inclusive
```

```python
data.head()
```

```python
# My custom code to divide to bins

divideToBins = data["Cholesterol"].max() - data["Cholesterol"].min()
divideToBins /= 10 # 10 is the no.of bins
start = data["Cholesterol"].min()
end = data["Cholesterol"].max()
i = 0
# print(start, end, divideToBins)
while start <= end:
  print(f"Bin {i} : {start} - {start + divideToBins}")
  i += 1
  start += divideToBins
```
